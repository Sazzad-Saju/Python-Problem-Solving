```python
# Proj01: ComplexNumber

# Create a complexNumber class that
# has a constructor with arg read_part = 0, img_part = 0
# overloaded add method that adds reat_part of two complex number and as well as add img_part
# create string representation 

#THIS COMPLEXNUMBER CLASS IS ALREADY DEFINED IN LIBRARY FUNCTIONS
# HERE IS REDEFINED
# Also presents the usages of the library function
```

```python
# Add Two complex number
# Redefined ComplexNumber class

class complexNumber:
    def __init__(self,real_part=0,img_part=0):
        self.real_part = real_part
        self.img_part = img_part
    def __str__(self):
        return "({0}+{1}j)".format(self.real_part,self.img_part)
    def __add__(self,other):
        real_part = self.real_part + other.real_part
        img_part = self.img_part + other.img_part
        return complexNumber(real_part,img_part)
        

cxObj1 = complexNumber(40,50)
cxObj2 = complexNumber(20,10)
cxObj3 = cxObj1 + cxObj2
print("{0} + {1} = {2}".format(cxObj1,cxObj2,cxObj3))

# With library function
a = 40 + 50j
b = 20 + 10j
c = a+b
print("{0} + {1} = {2}".format(a,b,c))

# create and delete a new attribute
cxObj3.new_attri = 80
print(cxObj3.new_attri)
del cxObj3.new_attri
# print(cxObj3.new_part)    #AttributeError
```

```python
# Proj04: Custom Iteration-Power of Two

# custom iterator: Power of Two upto limit
# Create a class with method __iter__() and __next__()
# can be modified without limitness
```

```python
# custom iterator: Power of Two upto limit
class powOfTwo:
    def __init__(self,max=0):
        self.max = max
    def __iter__(self):
        self.n = 0
        self.maxResult = 0
        return self
    def __next__(self):
        if self.maxResult <= self.max:
            result = 2 ** self.n
            self.maxResult += result
            self.n += 1
            return result
        else:
            raise StopIteration

a = powOfTwo(20)
i = iter(a)
print(next(i))
print(next(i))
print(next(i))
print(next(i))
print(next(i))
# print(next(i))  # StopIteration error when max = 20
```

```python
# Proj05: Person-Student Inheritance

# A inheritance: Student inherited from Person
# Person have first_name, last_name and age with method about()
# Student have extra dept and sid attribute with method Profile()
```

```python
# Inheritance
class Person:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def about(self):
        print("Name: {0} {1} \nAge: {2}".format(self.first_name, self.last_name, self.age))


class Student(Person):
    def __init__(self, first_name, last_name, age, dept, sid):
        super().__init__(first_name, last_name, age)
        self.dept = dept
        self.sid = sid

    def profile(self):
        super().about()
        print("Dept: ", self.dept)
        print("ID: ", self.sid)


print("Person1: ")
p1 = Person("Shanta", "Jannat", 26)
p1.about()
print("\nStudent1: ")
stnt1 = Student("Sazzad", "Saju", 27, "CSE", 1602049)
stnt1.profile()
# stnt1.about() can also be called
```

```python
# Proj06: Person-Student-Coder Multi Inheritance

# Multiple Inheritance
# Person and Student are different classes
# Coder inherit from Student and Person (Multiple Inheritance)
# But Here: 
# Coder inherit from Student where Student was inherited from Person hence-
# So Coder is inherited from both (Multiple Inheritance) 
```

```python
# Student have common with Person, Coder have common with Student

class Person:                   
    def __init__(self, first_name, last_name, age): 
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
    def about(self):
        print("Name: {0} {1} \nAge: {2}".format(self.first_name, self.last_name, self.age))
        
class Student(Person):                 
    def __init__(self,first_name, last_name, age, dept, sid):
        self.dept = dept
        self.sid = sid
        Person.__init__(self,first_name, last_name, age)
    def profile(self):
        Person.about(self)
        print("Dept: ", self.dept)
        print("ID: ", self.sid)
        
class Coder(Student):         
    def __init__(self, first_name, last_name, age, dept, sid, sector, profi): 
        self.sector = sector
        self.profi = profi
        Student.__init__(self,first_name, last_name, age, dept, sid)
    def info(self):
        Student.profile(self)
        print("Sector: ", self.sector)
        print("Proficiency: ", self.profi)
        
cr1 = Coder('Sazzad', 'Saju', 27, "CSE", 1602049, "Web development", "JavaScript")
cr1.info()
print()
st1 = Student('Shanta', 'Jannat', 26, "Medi", 150207)
st1.profile()
print()
pn1 = Person('Kutub', 'Ali', 65)
pn1.about()
```

```python
# Proj07: Simple Calculator

# input keyword
# calling function from function
# simple calculator can perform addition, substruction, multiplication and division
# Show menu to choose from
# raise WrongInputError if wrong input is chosen
```

```python

class WrongInputError(ValueError): pass 

def myAddition(x,y):
    print("{} + {} = {}".format(x,y,x+y)); return x+y
def mySubstraction(x,y):
    print("{} - {} = {}".format(x,y,x-y)); return x-y
def myMultiplication(x,y):
    print("{} * {} = {}".format(x,y,x*y)); return x*y
def myDivision(x,y):
    if y==0:
        raise ZeroDivisionError
    print("{} / {} = {}".format(x,y,x/y)); return x/y
    
def myMenu():
    print("Main Menu")
    print("1 > Addition")
    print("2 > Substraction")
    print("3 > Multiplication")
    print("4 > Division")
    choice = int(input("Please enter your option number: "))
    if choice>4 or choice<0: raise WrongInputError("Input number is wrong")
    return choice
def calculation():
    ch = myMenu()
    num1 = int(input("Please enter the first number: "))
    num2 = int(input("Plase enter the second number: "))
    if ch ==1:
        result = myAddition(num1,num2)
    elif ch == 2:
        result = mySubtraction(num1,num2)
    elif ch == 3:
        result = myMultiplication(num1,num2)
    elif ch == 4:
        result = myDivision(num1,num2)
    print("So result: ", result)

calculation()
```

```python
# Proj09: Operator Overloading for two Point

# Operator overloading
# overload add so it add two points: (4,5) + (3,6) = (7,11)
# overload less than op. to compare two points: if pow(4,2) + pow(5,2) < pow(3,2) + pow(6,2)
```

```python
class myPoint():
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
    def __str__(self):
        return "({0},{1})".format(self.x,self.y)
    # overloading + operator
    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y
        return myPoint(x,y)
    # overloading < operator
    def __lt__(self,other):
        self.mag = (self.x ** 2) + (self.y **2)
        other.mag = (other.x ** 2) + (other.y ** 2)
        return self.mag<other.mag

p1 = myPoint(4,5)
p2 = myPoint(3,6)
print(p1)
print(p2)
print()
print(p1+p2)
print(p1<p2)
```

```python
# Proj10: Division Decorator

# Python decorator
# a simple function go_divide, divide a number by another
# my_smart_div tries to modify the go_divide function to some extent i.e. raise ZeroDivisionError when
#occur and telling which one divided by what
```

```python
def my_smart_div(func):
    def inner_func(x,y):
        print("I an dividing ",x," by ",y)
        if y == 0:
            print("Oops! Divide by zero not possible")
            return
        return func(x,y)
    return inner_func


@my_smart_div
def go_divide(a,b):
    return a/b

print(go_divide(20,2))
print(go_divide(10,0))
```

```python
# Proj11: Sensoring Sentance

# Take a long string and a container of string. The long string can contains some of the container str
# Mute (replace with *) the matches in the long string (Sensoring)
# Implimented with Regex
```

```python
# Replace all matches
import re
badSent = "Sexy is a bad word in our case"
badWords = ['sexy','bad']
for i in range(len(badWords)):
    regex = re.compile(badWords[i],re.IGNORECASE)
    replaceWith = len(badWords[i]) * '*'
    sensorSent = regex.sub(replaceWith,badSent)
    badSent = sensorSent
print(sensorSent)
```

```python
# Proj13: Transpose Matrix

# Transpose a matrix
ex: 
# [1 4]
# [2 3]  -   [1 2 5 7]
# [5 6]  -   [4 3 6 5]
# [7 5]
```

```python
# Transposing a matrix using list comprehension
# example1
matrix = [[1,4], [2,3], [5,6], [7,5]]
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
#example2
matrix = [[1,4,2], [2,3,1], [5,6,3]]
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
```

```python
# Proj14: Recursion for Factorial

# Find factorial
# use recursion
```

```python
# recursion
# get factorial

def fact(n):
    if n <= 1:
        return 1;
    return n * fact(n-1)

num = int(input("Input a number: "))
print("Factorial of 5 is ", fact(num))
```

```python
# Proj15: Temperature Converter

# Use of Property for easy to debug
# raise error if wrong input is pass through constructor
# raise error if wrong input is set thourgh property
# use getter property to show temperature
# Temperature converter celcius to Fahrenheit
```

```python
# Use of Property
class tempConverter:
    def __init__(self,temperature = 0):
        if type(temperature) != int and type(temperature) != float:
            raise TypeError("Temperature must be a number")
        elif temperature < -273:
            raise ValueError("Temperature below -273 cannot convert")
        self._temperature = temperature
    def conv2Fahrenheit(self):
        return (self._temperature *1.8) + 32
    def get_temperature(self):
        return self._temperature
    def set_temperature(self,value):
        if type(value) != int and type(value) != float:
            raise TypeError("Temperature must be a number")
        elif value < -273:
            raise ValueError("Temperature below -273 cannot convert")
        self._temperature = value
    temperature = property(get_temperature, set_temperature)
    
# ValueError through constructor
# c = tempConverter(-300)

# ValueError through set property
# c = tempConverter(-30)
# c.temperature = -300

# TypeError through set property
# c = tempConverter(-30)
# c.temperature = 'c'

# TypeError through constructor
# c = tempConverter('c')

c = tempConverter(37.7778)
print(c.conv2Fahrenheit())
```
